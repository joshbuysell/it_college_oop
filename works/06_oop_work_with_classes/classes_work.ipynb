### 2. –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –ø—Ä–æ—Å—Ç–∏—Ö –¥–∞–Ω–∏—Ö –∑ –æ–±'—î–∫—Ç–∞–º–∏
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –ø–µ—Ä–µ–π–º–µ–Ω—É–π—Ç–µ –≤—Å—ñ –æ–±'—î–∫—Ç–∏ a1-a3 —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å —ó—Ö

```python
# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –ø—Ä–æ—Å—Ç–∏—Ö —Å—Ç—Ä—É–∫—Ç—É—Ä –¥–∞–Ω–∏—Ö
animals = ["–†–∏–±–∫–∏", "–ö—ñ—Ç", "–°–æ–±–∞–∫–∞"]
name = ["–ù–µ–æ–Ω–∏", "–ú—É—Ä–∫–∞", "–†–µ–∫—Å"]
age = [1, 5, 2]

print(f"–í –º–µ–Ω–µ —î —Ç–≤–∞—Ä–∏–Ω–∫–∞ {animals[1]} —è–∫—É –∑–≤–∞—Ç–∏ {name[1]} —è–∫—ñ {age[1]} —Ä–æ–∫—ñ–≤")

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –∫–ª–∞—Å—É
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"–í –º–µ–Ω–µ —î —Ç–≤–∞—Ä–∏–Ω–∫–∞ {animals} —è–∫—É –∑–≤–∞—Ç–∏ {name} —è–∫—ñ {age} —Ä–æ–∫—ñ–≤")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass == None else mass

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç—ñ–≤
fish = MyAnimals("–†–∏–±–∫–∏", "–ù–µ–æ–Ω–∏", 1)
cat = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5, 10)
dog = MyAnimals(animals="–°–æ–±–∞–∫–∞", name="–†–µ–∫—Å", age=2)

print(f"\n–û–±'—î–∫—Ç–∏ —Å—Ç–≤–æ—Ä–µ–Ω—ñ:")
print(f"1. {fish.animals} - {fish.name}")
print(f"2. {cat.animals} - {cat.name}")
print(f"3. {dog.animals} - {dog.name}")

# –ó–º—ñ–Ω–∞ –∞—Ç—Ä–∏–±—É—Ç—ñ–≤
cat.name = "–ú–∞—à–∫–∞"
print(f"\n–Ø –ø–µ—Ä–µ—ñ–º–µ–Ω—É–≤–∞–≤ —Å–≤–æ—é —Ç–≤–∞—Ä–∏–Ω–∫—É –Ω–∞ {cat.name}, –≤–æ–Ω–∞ –≤–∞–∂–∏—Ç—å {cat.mass} –∫–≥")
print(f"–ú–æ—è —Ç–≤–∞—Ä–∏–Ω–∫–∞ {fish.name}, –≤–æ–Ω–∞ –≤–∞–∂–∏—Ç—å {fish.mass} –∫–≥")
```

### 3. –ú–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –∑–∞—Å—Ç–æ—Å—É–π—Ç–µ –º–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É –¥–æ –≤—Å—ñ—Ö –æ–±'—î–∫—Ç—ñ–≤

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass == None else mass
    
    def speak(self):
        """–ú–µ—Ç–æ–¥ –ø–æ–≤–µ—Ä—Ç–∞—î –∑–≤—É–∫ —Ç–≤–∞—Ä–∏–Ω–∫–∏"""
        if self.animals == "–ö—ñ—Ç":
            return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –≥–æ–≤–æ—Ä–∏—Ç—å –ú—è—É—É—É"
        elif self.animals == "–°–æ–±–∞–∫–∞":
            return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –≥–æ–≤–æ—Ä–∏—Ç—å –ì–∞–≤–≤"
        else:
            return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –Ω–µ —Ä–æ–∑–º–æ–≤–ª—è—î"
    
    def add_one_year(self):
        """–ú–µ—Ç–æ–¥ –¥–æ–¥–∞—î –æ–¥–∏–Ω —Ä—ñ–∫ –¥–æ –≤—ñ–∫—É"""
        self.age += 1
        return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑–∞ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"
    
    def info(self):
        """–î–æ–¥–∞—Ç–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ –¥–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó"""
        return f"{self.animals} {self.name}, –≤—ñ–∫: {self.age} —Ä–æ–∫—ñ–≤, –≤–∞–≥–∞: {self.mass} –∫–≥"

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±'—î–∫—Ç—ñ–≤
a1 = MyAnimals("–†–∏–±–∫–∏", "–ù–µ–æ–Ω–∏", 1)
a2 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5, 10)
a3 = MyAnimals(animals="–°–æ–±–∞–∫–∞", name="–†–µ–∫—Å", age=2)

# –ó–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è –º–µ—Ç–æ–¥—ñ–≤
print(f"–¢–≤–∞—Ä–∏–Ω–∫–∞ {a1.animals}:", a1.speak())
print(f"–¢–≤–∞—Ä–∏–Ω–∫–∞ {a2.animals}:", a2.speak())
print(f"–¢–≤–∞—Ä–∏–Ω–∫–∞ {a3.animals}:", a3.speak())

print(f"\n{a1.animals} –∂–∏–≤–µ –∑—ñ –º–Ω–æ—é {a1.age} —Ä–æ–∫—ñ–≤")
print(a1.add_one_year())
print(f"{a1.animals} –∂–∏–≤–µ –∑—ñ –º–Ω–æ—é {a1.age} —Ä–æ–∫—ñ–≤")

# –î–æ–¥–∞—Ç–∫–æ–≤–µ –∑–∞–≤–¥–∞–Ω–Ω—è - –≤–∏–≤–µ–¥–µ–Ω–Ω—è —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó –ø—Ä–æ –≤—Å—ñ—Ö —Ç–≤–∞—Ä–∏–Ω
print("\n--- –Ü–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—è –ø—Ä–æ –≤—Å—ñ—Ö —Ç–≤–∞—Ä–∏–Ω ---")
for animal in [a1, a2, a3]:
    print(animal.info())
```

### 4. –ó–º—ñ–Ω–Ω—ñ –∫–ª–∞—Å—É (–≥–ª–æ–±–∞–ª—å–Ω—ñ –∑–º—ñ–Ω–Ω—ñ –Ω–∞–¥ –≤—Å—ñ–º–∞ –æ–±'—î–∫—Ç–∞–º–∏)
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –∑–∞–¥–∞–π—Ç–µ –¥–µ–∫—ñ–ª—å–∫–∞ –æ–±'—î–∫—Ç—ñ–≤ –≤ —Ü–∏–∫–ª—ñ —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –≥–ª–æ–±–∞–ª—å–Ω—É –∑–º—ñ–Ω–Ω—É

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass == None else mass
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

print(f"–í –º–µ–Ω–µ –∑–∞—Ä–∞–∑ —î —Ç–≤–∞—Ä–∏–Ω–∫–∏ {list(MyAnimals.total_animal_types)}, —ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å {MyAnimals.total_animals}")

a1 = MyAnimals("–†–∏–±–∫–∏", "–ù–µ–æ–Ω–∏", 1)
MyAnimals("–†–∏–±–∫–∏", "–ì—É–ø—ñ—ó", 1)
MyAnimals("–†–∏–±–∫–∏", "–°–æ–º—ñ–∫–∏", 1)
MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5, 10)

print(f"–í –º–µ–Ω–µ –∑–∞—Ä–∞–∑ —î —Ç–≤–∞—Ä–∏–Ω–∫–∏ {list(MyAnimals.total_animal_types)}, —ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å {MyAnimals.total_animals}")

print(f"\n–Ø –¥–æ—Å—Ç—É–ø–∞—é—Å—å –¥–æ –∑–º—ñ–Ω–Ω–æ—ó –∫–ª–∞—Å—É –∑ –æ–±'—î–∫—Ç–∞: {a1.total_animal_types}")
a1.total_animal_types = {"–°–æ–±–∞–∫–∞"}  # –Ø –∑—Ä–æ–±–∏–≤ –∑–º—ñ–Ω–Ω—É –∫–ª–∞—Å—É –∞—Ç—Ä–∏–±—É—Ç–æ–º –æ–±'—î–∫—Ç—É
print(f"–Ø –¥–æ—Å—Ç—É–ø–∞—é—Å—å –¥–æ –∑–º—ñ–Ω–Ω–æ—ó –∫–ª–∞—Å—É –∑ –æ–±'—î–∫—Ç–∞ {a1.total_animal_types} —Ç–∞ –∑ –∫–ª–∞—Å—É {MyAnimals.total_animal_types}")

# –î–æ–¥–∞—Ç–∫–æ–≤–µ –∑–∞–≤–¥–∞–Ω–Ω—è - —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–≤–∞—Ä–∏–Ω —É —Ü–∏–∫–ª—ñ
print("\n--- –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–≤–∞—Ä–∏–Ω —É —Ü–∏–∫–ª—ñ ---")
animals_data = [
    ("–ü–∞–ø—É–≥–∞", "–ö–µ—à–∞", 3),
    ("–•–æ–º'—è–∫", "–ü—É—Ö", 1),
    ("–ß–µ—Ä–µ–ø–∞—Ö–∞", "–¢–æ—Ä—Ç–∏–ª–∞", 50)
]

for animal, name, age in animals_data:
    MyAnimals(animal, name, age)

print(f"\n–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ç–≤–∞—Ä–∏–Ω: {MyAnimals.total_animals}")
print(f"–¢–∏–ø–∏ —Ç–≤–∞—Ä–∏–Ω: {list(MyAnimals.total_animal_types)}")
```

### 5. –°—Ç–∞—Ç–∏—á–Ω—ñ –º–µ—Ç–æ–¥–∏ —Ç–∞ –∑–º—ñ–Ω–Ω–∞ self
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –≤–∏–∫–ª–∏—á—Ç–µ —Å—Ç–∞—Ç–∏—á–Ω–∏–π –º–µ—Ç–æ–¥ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –æ–±'—î–∫—Ç–∞

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        """–ó–≤–∏—á–∞–π–Ω–∏–π –º–µ—Ç–æ–¥, –ø–æ—Ç—Ä–µ–±—É—î self"""
        self.age += 1
        return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑–∞ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"

    @staticmethod
    def call_pet():
        """–°—Ç–∞—Ç–∏—á–Ω–∏–π –º–µ—Ç–æ–¥, –Ω–µ –ø–æ—Ç—Ä–µ–±—É—î self"""
        return "–¢–≤–∞—Ä–∏–Ω–∫–∞ –±—ñ–∂–∏—Ç—å –¥–æ –Ω–∞—Å"
    
    @staticmethod
    def pet_advice():
        """–î–æ–¥–∞—Ç–∫–æ–≤–∏–π —Å—Ç–∞—Ç–∏—á–Ω–∏–π –º–µ—Ç–æ–¥"""
        return "–ù–µ –∑–∞–±—É–≤–∞–π—Ç–µ –≥–æ–¥—É–≤–∞—Ç–∏ —Å–≤–æ—ó—Ö —Ç–≤–∞—Ä–∏–Ω —â–æ–¥–Ω—è!"

a1 = MyAnimals("–†–∏–±–∫–∏", "–ì—É–ø—ñ—ó", 1)
a2 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)

print(f"–ú–∏ –∫–ª–∏—á–µ–º–æ –Ω–∞—à—É —Ç–≤–∞—Ä–∏–Ω–∫—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Å—Ç–∞—Ç–∏—á–Ω–æ–≥–æ –º–µ—Ç–æ–¥—É:")
print(f"  - –∑ –æ–±'—î–∫—Ç—É: {a1.call_pet()}")
print(f"  - –∑ –∫–ª–∞—Å—É: {MyAnimals.call_pet()}")

print(f"\n–í–∏–∫–ª–∏–∫–∞—é –º–µ—Ç–æ–¥ add_one_year:")
print(f"  - –∑ –æ–±'—î–∫—Ç—É: {a1.add_one_year()}")
print(f"  - –∑ –∫–ª–∞—Å—É (–ø–µ—Ä–µ–¥–∞—é—á–∏ self –≤—Ä—É—á–Ω—É): {MyAnimals.add_one_year(a1)}")

# –î–æ–¥–∞—Ç–∫–æ–≤–µ –∑–∞–≤–¥–∞–Ω–Ω—è
print(f"\n–ü–æ—Ä–∞–¥–∞ –¥–ª—è –≤—Å—ñ—Ö: {MyAnimals.pet_advice()}")
print(f"–ü–æ—Ä–∞–¥–∞ –≤—ñ–¥ –æ–±'—î–∫—Ç–∞: {a2.pet_advice()}")
```

### 6. –†–æ–±–æ—Ç–∞ –∑ properties
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –¥–æ–¥–∞–π—Ç–µ —â–µ –æ–¥–Ω—É property —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å —ó—ó

```python
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age, birth_year=None) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self.birth_year = birth_year if birth_year else datetime.today().year - age
    
    @property
    def live_until(self):
        """–†–æ–∑—Ä–∞—Ö–æ–≤—É—î —Ä—ñ–∫ –¥–æ —è–∫–æ–≥–æ –ø—Ä–æ–∂–∏–≤–µ —Ç–≤–∞—Ä–∏–Ω–∞"""
        return datetime.today().year - self.age + self.max_age
    
    @property
    def age_in_human_years(self):
        """–î–æ–¥–∞—Ç–∫–æ–≤–∞ property - –≤—ñ–∫ —É –ª—é–¥—Å—å–∫–∏—Ö —Ä–æ–∫–∞—Ö"""
        if self.animals == "–ö—ñ—Ç" or self.animals == "–°–æ–±–∞–∫–∞":
            return self.age * 7
        return self.age
    
    @property
    def description(self):
        """–ü–æ–≤–Ω–∏–π –æ–ø–∏—Å —Ç–≤–∞—Ä–∏–Ω–∏"""
        return f"{self.animals} {self.name}, –≤—ñ–∫: {self.age} —Ä–æ–∫—ñ–≤ (—É –ª—é–¥—Å—å–∫–∏—Ö —Ä–æ–∫–∞—Ö: {self.age_in_human_years})"
    
    def add_one_year(self):
        self.age += 1
        return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑–∞ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"

a1 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)
print(f"–ú—ñ–π {a1.animals} –±—É–¥–µ –∑—ñ –º–Ω–æ—é –¥–æ {a1.live_until} —Ä–æ–∫—É")
a1.add_one_year()
print(f"–ú—ñ–π {a1.animals} –±—É–¥–µ –∑—ñ –º–Ω–æ—é –¥–æ {a1.live_until} —Ä–æ–∫—É")

# –î–æ–¥–∞—Ç–∫–æ–≤—ñ –∑–∞–≤–¥–∞–Ω–Ω—è
print(f"\n–í—ñ–∫ —É –ª—é–¥—Å—å–∫–∏—Ö —Ä–æ–∫–∞—Ö: {a1.age_in_human_years}")
print(f"–ü–æ–≤–Ω–∏–π –æ–ø–∏—Å: {a1.description}")

a2 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 3)
print(f"\n{a2.description}")
print(f"–ü—Ä–æ–∂–∏–≤–µ –¥–æ {a2.live_until} —Ä–æ–∫—É")
```

### 7. –ú–µ—Ç–æ–¥–∏ –ö–ª–∞—Å—É (classmethod)
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –¥–æ–ø–∏—à—ñ—Ç—å –º–µ—Ç–æ–¥ –∫–ª–∞—Å—É —è–∫–∏–π –±—É–¥–µ –ø–µ—Ä–µ–≤—ñ—Ä—è—Ç–∏ —â–æ –∞—Ç—Ä–∏–±—É—Ç name –∑–∞–¥–∞–Ω–∏–π –∑ –≤–µ–ª–∏–∫–æ—ó –±—É–∫–≤–∏

```python
class MyAnimals:
    total_csv = 0
    total_objects = 0

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        """–°—Ç–≤–æ—Ä—é—î –æ–±'—î–∫—Ç –∑ CSV —Ä—è–¥–∫–∞"""
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)
    
    @classmethod
    def validate_name(cls, name: str) -> bool:
        """–ü–µ—Ä–µ–≤—ñ—Ä—è—î —á–∏ —ñ–º'—è –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –≤–µ–ª–∏–∫–æ—ó –±—É–∫–≤–∏"""
        if name and name[0].isupper():
            return True
        return False
    
    @classmethod
    def create_with_validation(cls, animals, name, age):
        """–°—Ç–≤–æ—Ä—é—î –æ–±'—î–∫—Ç –∑ –ø–µ—Ä–µ–≤—ñ—Ä–∫–æ—é —ñ–º–µ–Ω—ñ"""
        if not cls.validate_name(name):
            print(f"–ü–æ–ø–µ—Ä–µ–¥–∂–µ–Ω–Ω—è: —ñ–º'—è '{name}' –Ω–µ –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –≤–µ–ª–∏–∫–æ—ó –±—É–∫–≤–∏!")
            name = name.capitalize()
            print(f"–í–∏–ø—Ä–∞–≤–ª–µ–Ω–æ –Ω–∞: '{name}'")
        return cls(animals, name, age)

a1 = MyAnimals.from_csv("–ö—ñ—Ç,–ú—É—Ä–∫–∞,5")
a2 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 2)

print(f"–ú—ñ–π {a1.animals}, –∑–≤–∞—Ç–∏ {a1.name}, –º–∞—î {a1.age} —Ä–æ–∫—ñ–≤")
print(f"–ú—ñ–π {a2.animals}, –∑–≤–∞—Ç–∏ {a2.name}, –º–∞—î {a2.age} —Ä–æ–∫—ñ–≤")
print(f"–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–±'—î–∫—Ç—ñ–≤ {MyAnimals.total_objects} –∞ —Å—Ç–≤–æ—Ä–µ–Ω–∏—Ö –∑ csv —Ñ–æ—Ä–º–∞—Ç—É {MyAnimals.total_csv}")

# –î–æ–¥–∞—Ç–∫–æ–≤–µ –∑–∞–≤–¥–∞–Ω–Ω—è - –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ —ñ–º–µ–Ω
print("\n--- –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ —ñ–º–µ–Ω ---")
print(f"–Ü–º'—è '–ú—É—Ä–∫–∞' –≤–∞–ª—ñ–¥–Ω–µ: {MyAnimals.validate_name('–ú—É—Ä–∫–∞')}")
print(f"–Ü–º'—è '—Ä–µ–∫—Å' –≤–∞–ª—ñ–¥–Ω–µ: {MyAnimals.validate_name('—Ä–µ–∫—Å')}")

print("\n--- –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑ –≤–∞–ª—ñ–¥–∞—Ü—ñ—î—é ---")
a3 = MyAnimals.create_with_validation("–ü–∞–ø—É–≥–∞", "–∫–µ—à–∞", 3)
print(f"–°—Ç–≤–æ—Ä–µ–Ω–æ: {a3.animals} {a3.name}")
```

### 8. –ü—Ä–∏–≤–∞—Ç–Ω—ñ –∞—Ç—Ä–∏–±—É—Ç–∏ —Ç–∞ –º–µ—Ç–æ–¥–∏
**–ó–∞–≤–¥–∞–Ω–Ω—è:** —Å—Ç–≤–æ—Ä—ñ—Ç—å –¥–æ–¥–∞—Ç–∫–æ–≤–∏–π –ø—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç —Ç–∞ –º–µ—Ç–æ–¥ –¥–ª—è –¥–æ—Å—Ç—É–ø—É –¥–æ –Ω—å–æ–≥–æ

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self._secret_phrase = "–¶–µ –ø—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç (—è–∫ —è –±—É–¥—É –∫–ª–∏–∫–∞—Ç–∏ —Å–≤–æ—é —Ç–≤–∞—Ä–∏–Ω–∫—É)"
        self.__secret_strong = "–¶–µ–π –∞—Ç—Ä–∏–±—É—Ç —Ç–æ—á–Ω–æ –Ω–µ –º–æ–∂–Ω–∞ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ –∑–∑–æ–≤–Ω—ñ!!"
        self.__id = id(self)  # –î–æ–¥–∞—Ç–∫–æ–≤–∏–π –ø—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç
    
    @property
    def secret_phrase(self):
        return f"–ú–∏ –¥–æ—Å—Ç—É–ø–∞—î–º–æ—Å—å –¥–æ –ø—Ä–∏–≤–∞—Ç–Ω–æ–≥–æ –∞—Ç—Ä–∏–±—É—Ç—É –∑ property: {self._secret_phrase}"
    
    @property
    def secret_strong(self):
        return f"–ú–∏ –¥–æ—Å—Ç—É–ø–∞—î–º–æ—Å—å –¥–æ –ø—Ä–∏–≤–∞—Ç–Ω–æ–≥–æ –∞—Ç—Ä–∏–±—É—Ç—É –∑ property: {self.__secret_strong}"
    
    @property
    def object_id(self):
        """–î–æ–¥–∞—Ç–∫–æ–≤–∏–π property –¥–ª—è –¥–æ—Å—Ç—É–ø—É –¥–æ –ø—Ä–∏–≤–∞—Ç–Ω–æ–≥–æ ID"""
        return f"ID –æ–±'—î–∫—Ç–∞: {self.__id}"
    
    def _secret_method(self):
        return "–ü—Ä–∏–≤–∞—Ç–Ω–∏–π –º–µ—Ç–æ–¥, –∫—Ä–∞—â–µ –π–æ–≥–æ –Ω–µ —Ä—É—Ö–∞—Ç–∏, –≤—ñ–Ω —â–æ—Å—å —Ä–æ–±–∏—Ç—å –ø–æ—Ç—Ä—ñ–±–Ω–µ –≤—Å–µ—Ä–µ–¥–∏–Ω—ñ –∫–ª–∞—Å—É"

    def __do_not_touch_it(self):
        return "–¶–µ –Ω–µ–º–æ–∂–ª–∏–≤–æ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ –∑–∑–æ–≤–Ω—ñ"
    
    def public_method_using_private(self):
        """–ü—É–±–ª—ñ—á–Ω–∏–π –º–µ—Ç–æ–¥ —â–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –ø—Ä–∏–≤–∞—Ç–Ω–∏–π"""
        return f"–ü—É–±–ª—ñ—á–Ω–∏–π –º–µ—Ç–æ–¥ –≤–∏–∫–ª–∏–∫–∞–≤: {self.__do_not_touch_it()}"

a1 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 2)

print("--- –î–æ—Å—Ç—É–ø –¥–æ –ø—Ä–∏–≤–∞—Ç–Ω–∏—Ö –∞—Ç—Ä–∏–±—É—Ç—ñ–≤ ---")
print(a1._secret_phrase)  # –¢–∞–∫ —Ä–æ–±–∏—Ç–∏ –Ω–µ —Ç—Ä–µ–±–∞!
a1._secret_phrase = "–Ø –∑–º—ñ–Ω–∏–≤ –ø—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç"  # –¢–∞–∫ —Ä–æ–±–∏—Ç–∏ –Ω–µ —Ç—Ä–µ–±–∞!
print(a1._secret_phrase)

# print(a1.__secret_strong)  # –¶–µ –≤–∏–¥–∞—Å—Ç—å –ø–æ–º–∏–ª–∫—É!
print(a1._MyAnimals__secret_strong)  # –¢–∞–∫ —Ç–∞–∫–æ–∂ —Ä–æ–±–∏—Ç–∏ –Ω–µ –º–æ–∂–Ω–∞!

print("\n--- –í–∏–≤–æ–¥–∏–º–æ –ø—Ä–∏–≤–∞—Ç–Ω—ñ –∞—Ç—Ä–∏–±—É—Ç–∏ —á–µ—Ä–µ–∑ property ---")
print(a1.secret_phrase)
print(a1.secret_strong)
print(a1.object_id)  # –î–æ–¥–∞—Ç–∫–æ–≤–µ –∑–∞–≤–¥–∞–Ω–Ω—è

print("\n--- –†–æ–±–æ—Ç–∞ –∑ –ø—Ä–∏–≤–∞—Ç–Ω–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏ ---")
print(a1._secret_method())
# a1.__do_not_touch_it()  # –¶–µ –≤–∏–¥–∞—Å—Ç—å –ø–æ–º–∏–ª–∫—É!
print(a1.public_method_using_private())  # –ü—Ä–∞–≤–∏–ª—å–Ω–∏–π —Å–ø–æ—Å—ñ–± –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
```

### 9. –ú–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –¥–æ–¥–∞–π—Ç–µ –¥–æ–¥–∞—Ç–∫–æ–≤—ñ –º–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        """–û—Ñ—ñ—Ü—ñ–π–Ω–µ —Å—Ç—Ä–æ–∫–æ–≤–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞"""
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        """–î—Ä—É–∂–Ω—î —Å—Ç—Ä–æ–∫–æ–≤–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—è –æ–±'—î–∫—Ç–∞"""
        return f"–¶–µ —î –æ–±'—î–∫—Ç {self.__repr__()}"
    
    def __eq__(self, other) -> bool:
        """–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–∞ —Ä—ñ–≤–Ω—ñ—Å—Ç—å"""
        if isinstance(other, MyAnimals):
            return self.name == other.name and self.animals == other.animals
        return False
    
    def __hash__(self) -> int:
        """–•–µ—à –æ–±'—î–∫—Ç–∞"""
        return hash((self.animals, self.name))

a1 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 2)
a2 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 3)
a3 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)

print("--- –ú–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏ __str__ —Ç–∞ __repr__ ---")
print(a1)
print(repr(a1))
print(str(a1))

print("\n--- –ú–∞–≥—ñ—á–Ω–∏–π –º–µ—Ç–æ–¥ __eq__ ---")
print(f"a1 == a2: {a1 == a2}")
print(f"a1 == a3: {a1 == a3}")

print("\n--- –ú–∞–≥—ñ—á–Ω–∏–π –º–µ—Ç–æ–¥ __hash__ ---")
print(f"hash(a1): {hash(a1)}")
print(f"hash(a2): {hash(a2)}")
print(f"hash(a3): {hash(a3)}")
```

### 10. –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω—ñ –º–µ—Ç–æ–¥–∏
**–ó–∞–≤–¥–∞–Ω–Ω—è:** –¥–æ–¥–∞–π—Ç–µ –¥–æ–¥–∞—Ç–∫–æ–≤–∏–π –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–∏–π –º–µ—Ç–æ–¥

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        """–í—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è"""
        if isinstance(obj, MyAnimals):
            return f"–ú—ñ–π {self.name} –ø–æ–∫—É—Å–∞–≤ {obj.name}"
        elif isinstance(obj, int):
            self.age += obj
            return f"–ú—ñ–π {self.name} –ø–æ—Å—Ç–∞—Ä—ñ–≤ –Ω–∞ {obj} —Ä–æ–∫—ñ–≤, —Ç–µ–ø–µ—Ä –π–æ–º—É {self.age}"
        return NotImplementedError
    
    def __gt__(self, obj):
        """–ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –±—ñ–ª—å—à–µ"""
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"–°—Ç–∞—Ä—à–∏–º —î {self.name}"
            return f"{self.name} –Ω–µ —î —Å—Ç–∞—Ä—à–∏–º –∑–∞ {obj.name}"
        return NotImplementedError

    def __len__(self) -> int:
        """–î–æ–≤–∂–∏–Ω–∞ —ñ–º–µ–Ω—ñ"""
        print("–ú–∏ –∑–∞–π—à–ª–∏ –≤ –º–µ—Ç–æ–¥ len, —Ç–∞ –∑–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–≤–∂–∏–Ω—É –Ü–º–µ–Ω—ñ —Ç–≤–∞—Ä–∏–Ω–∫–∏")
        return len(self.name)
    
    def __add__(self, obj):
        """–î–æ–¥–∞—Ç–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ - –¥–æ–¥–∞–≤–∞–Ω–Ω—è"""
        if isinstance(obj, MyAnimals):
            return f"{self.name} —Ç–∞ {obj.name} —Å—Ç–∞–ª–∏ –¥—Ä—É–∑—è–º–∏"
        elif isinstance(obj, int):
            new_age = self.age + obj
            return f"{self.name} –ø–æ–º–æ–ª–æ–¥—à–∞–≤ –Ω–∞ {obj} —Ä–æ–∫—ñ–≤, —Ç–µ–ø–µ—Ä –π–æ–º—É {new_age}"
        return NotImplementedError
    
    def __mul__(self, obj):
        """–î–æ–¥–∞—Ç–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ - –º–Ω–æ–∂–µ–Ω–Ω—è"""
        if isinstance(obj, int):
            return f"{self.name} * {obj} = {'üêæ' * obj}"
        return NotImplementedError

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è –±–∞–∑–æ–≤–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π
print("--- –ë–∞–∑–æ–≤—ñ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó ---")
a = 2
b = 5
c = "C"
print(f"–ü—Ä–æ—Å—Ç—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó: {a+b}, {a - 1}, {a * b}, {str(a) + c}")
print(f"–ß–µ—Ä–µ–∑ –º–∞–≥—ñ—á–Ω—ñ –º–µ—Ç–æ–¥–∏: {a.__add__(b)}, {a.__sub__(1)}, {a.__mul__(b)}")

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è –∑ –∫–ª–∞—Å–æ–º
print("\n--- –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω—ñ –º–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É ---")
a = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 2)
b = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∑–∏–∫", 4)

print(a - b)
print(a - 1)
print(a - float(1))

print("\n--- –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è ---")
print(a > b)
print(b > a)

print("\n--- –î–æ–≤–∂–∏–Ω–∞ ---")
print(f"–î–æ–≤–∂–∏–Ω–∞ —ñ–º–µ–Ω—ñ: {len(a)}")

print("\n--- –î–æ–¥–∞—Ç–∫–æ–≤—ñ –º–µ—Ç–æ–¥–∏ ---")
print(a + b)
print(a + 3)
print(a * 5)
```
