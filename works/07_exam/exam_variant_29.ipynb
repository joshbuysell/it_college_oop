## Промпт для AI

Створи клас StringLengthCalculator, який обчислює довжину рядка без пробілів.

## Реалізація класу

```python
class StringLengthCalculator:
    """Клас для обчислення довжини рядка без пробілів"""
    
    def __init__(self, text: str):
        """Ініціалізація класу з рядком
        
        Args:
            text (str): Вхідний рядок для обробки
        """
        self._text = text
        self._length = None
    
    def calculate(self):
        """Метод для обчислення довжини рядка без пробілів"""
        # Видаляємо всі пробіли з рядка та обчислюємо довжину
        text_without_spaces = self._text.replace(" ", "")
        self._length = len(text_without_spaces)
    
    @property
    def length_without_spaces(self) -> int:
        """Property для отримання довжини рядка без пробілів
        
        Returns:
            int: Довжина рядка без пробілів
        """
        if self._length is None:
            self.calculate()
        return self._length
    
    def __str__(self) -> str:
        """Метод для виведення інформації про студента
        
        Returns:
            str: Інформація про студента
        """
        return "Ім'я студента: Віктор Цвик, Група: КН-210"


# Набори тестових даних
test_strings = [
    "Привіт світ",
    "Об'єктно орієнтоване програмування",
    "Python це цікаво"
]

# Створення об'єктів в циклі
calculators = []

for text in test_strings:
    calculator = StringLengthCalculator(text)
    calculators.append(calculator)

# Виведення результатів
for calculator in calculators:
    print(str(calculator))
    print(f"Вхідний рядок: '{calculator._text}'")
    print(f"Довжина без пробілів: {calculator.length_without_spaces}")
    print("-" * 50)
```
